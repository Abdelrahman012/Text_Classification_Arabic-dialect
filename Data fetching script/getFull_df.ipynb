```python
import json
import pandas as pd
import requests
DATA_PATH="..//Data//dialect_dataset.csv"
EXPORT_PATH = "..//Data//processed//final_df.pkl"
```

```python
##get list of lists in each list 1000 id as  a string to get the text data from API when send the request
def get_ids_lists(df):
    lists_data_1000=[]
    lst_1000=[]
    item=0
    for i,row in df.iterrows():
        lst_1000.append(str(row["id"]))
        if len(lst_1000)==1000:
            lists_data_1000.append(lst_1000)
            lst_1000=[]
            item+=1
        elif i==df.shape[0]-1:
            lists_data_1000.append(lst_1000)
    return lists_data_1000
```

```python
lists_data_1000=get_ids_lists(df)
len(lists_data_1000)
```

```python
def get_df_with_texts(lists_data_1000,df):
    lst_id=[]
    lst_text=[]
    n=0
    for lst in lists_data_1000:
        print(f"the iteration is: {n}")
        n+=1
        r = requests.post('https://recruitment.aimtechnologies.co/ai-tasks',json=lst, headers = {"content-type": "application/json"})
        tet=json.loads(r.text)
        for item in tet.items():
            lst_id.append(item[0])
            lst_text.append(item[1])
    print(len(lst_id),len(lst_text),len(df["dialect"].to_list()))
    final_df=pd.DataFrame({"id":lst_id,"texts":lst_text,"dialect":df["dialect"].to_list()})
    return final_df
```

```python
final_df=get_df_with_texts(lists_data_1000,df)
```
